```python
import pandas as pd

df = pd.read_csv(
    "sales_data.csv",
    encoding="latin1",
    engine="python"
)

print(df.shape)
df.head()
```

```python
# Dataset Information
print("Shape of Dataset:", df.shape)
print("\nColumn Names:")
print(df.columns)

print("\nFirst 5 Rows:")
display(df.head())
```

```python
# Check Missing Values
print(df.isnull().sum())
```

```python
# Summary Statistics
df.describe()
```

```python
# Convert Order Date into datetime format
df["Order Date"] = pd.to_datetime(df["Order Date"], dayfirst=True)

# Check data type
print(df["Order Date"].dtype)
```

```python
# Create Year-Month column
df["Year_Month"] = df["Order Date"].dt.to_period("M")

# Group data by month
monthly_sales = df.groupby("Year_Month")["Sales"].sum().reset_index()

# Convert Period to Timestamp
monthly_sales["Year_Month"] = monthly_sales["Year_Month"].dt.to_timestamp()

monthly_sales.head()
```

```python
plt.figure(figsize=(12,5))
plt.plot(monthly_sales["Year_Month"], monthly_sales["Sales"], marker="o")
plt.title("Monthly Sales Trend")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()
```

```python
# Create a numerical time index
monthly_sales["Month_Number"] = range(1, len(monthly_sales) + 1)

monthly_sales.head()
```

```python
# Features and Target
X = monthly_sales[["Month_Number"]]
y = monthly_sales["Sales"]
```

```python
# Train Model
model = LinearRegression()
model.fit(X, y)

# Predictions
monthly_sales["Predicted_Sales"] = model.predict(X)

monthly_sales.head()
```

```python
mae = mean_absolute_error(y, monthly_sales["Predicted_Sales"])
rmse = np.sqrt(mean_squared_error(y, monthly_sales["Predicted_Sales"]))
r2 = r2_score(y, monthly_sales["Predicted_Sales"])

print("MAE :", round(mae,2))
print("RMSE:", round(rmse,2))
print("R² Score:", round(r2,2))
```

```python
plt.figure(figsize=(12,5))

plt.plot(monthly_sales["Year_Month"], monthly_sales["Sales"],
         label="Actual Sales", marker="o")

plt.plot(monthly_sales["Year_Month"], monthly_sales["Predicted_Sales"],
         label="Predicted Sales", linestyle="--")

plt.title("Actual vs Predicted Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)
plt.xticks(rotation=45)

plt.show()
```

```python
# Predict next month's sales
next_month = pd.DataFrame({"Month_Number": [len(monthly_sales) + 1]})

forecast = model.predict(next_month)

print("Predicted Sales for Next Month: $", round(forecast[0], 2))
```
